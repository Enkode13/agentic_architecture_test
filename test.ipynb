```python
import semchunk
from sentence_transformers import SentenceTransformer

# 1. Extraction (PyMuPDF)
doc = pymupdf.open(source)
full_text = ""

for page in doc:
    # Use "blocks" to maintain a logical reading order
    blocks = page.get_text("blocks")
    for b in blocks:
        # b[4] is the text content
        full_text += b[4] + "\n\n"

full_text
```

```python
# 2. Semantic Chunking
# Using a tokenizer ensures chunks fit perfectly into your embedding model
encoder = SentenceTransformer('BAAI/bge-base-en-v1.5') 
tokenizer = encoder.tokenizer
```

```python
# semchunk handles the "overlap" and "context" logic internally
chunk_size = 400
chunker = semchunk.chunkerify(
    tokenizer,
    chunk_size, 
)

chunks = chunker(full_text)

print(f"Created {len(chunks)} semantic chunks.")
```
